```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/file.csv")
G = nx.Graph()

for _, row in df.iterrows():
    G.add_node(row['user'], type='user')
    G.add_node(row['file'], type='file')
    G.add_edge(row['user'], row['file'])

sub_nodes = list(G.nodes)[:100]
subgraph = G.subgraph(sub_nodes)
colors = ['blue' if G.nodes[n]['type'] == 'user' else 'green' for n in subgraph.nodes]

nx.draw(subgraph, with_labels=True, node_color=colors, node_size=200, font_size=6)
```
